```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = 'data/BTC-USD.csv'
```

```python
df = pd.read_csv(path, index_col='Date', parse_dates=True)
df.head()
```

```python
df.index
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,8))
df['2020']['Close'].plot()
df['2020']['Open'].plot(ls=':', alpha=0.8)
plt.ylabel('Price in USD')
plt.legend()
```

```python
plt.figure(figsize=(10,5))
df['2020-3']['Close'].plot()
df['2020-3']['Close'].resample('W').mean().plot(label='Moyenne par mois', lw=3,ls=':',alpha=0.8)
df['2020-3']['Close'].resample('W').mean().plot(label='Variation par semaine', lw=3,ls='--',alpha=0.8)
plt.title('Bitcoin Evolution')
plt.legend()
plt.show()
```
