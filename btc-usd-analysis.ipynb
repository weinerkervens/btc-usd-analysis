```python
# Useful Imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Function to import timeseries datasets

def import_df(name,index_col='Date'):
    ext = '.csv'
    root = 'data/'
    df = pd.read_csv(root+name+ext,index_col=index_col,parse_dates=True)
    return df
```

```python
# Importing BTC dataset

btc = import_df('BTC-USD')
btc.head()
```

```python
# Importing ETH dataset

eth = import_df('ETH-USD')
eth.head()
```

```python
# Importing HTG dataset

htg = import_df('USD-HTG')
htg.head()
```

```python
plt.figure(figsize=(15,8))
btc['2017-09':]['Close'].plot(label='BTC',lw=1, alpha=1)
eth['2017-09':]['Close'].plot(label='ETH',alpha=.5)
#htg['2015-09':]['Close'].plot(label='HTG',alpha=.4)
#btc['2015-09':]['Close'].max().plot(kind='scatter',label='BTC',ls='*',lw=1, alpha=1)
#plt.plot(x=)
plt.ylabel('Price in USD')
plt.legend()
```

```python
# Function to display evolution of currency

def plot_evol(name,data,year,feature):
    plt.figure(figsize=(15,5))
    data[year][feature].plot()
    data[year][feature].resample('M').mean().plot(label='Moyenne par mois', lw=3,alpha=0.8)
    data[year][feature].resample('M').std().plot(label='Variation par mois', lw=3,ls='--',alpha=0.8)
    data[year][feature].diff().plot(label='Volatilit',alpha=0.8)
    plt.title(f'Evolution du {name} en {year}\n')
    plt.ylabel('Prix en USD')
    plt.legend()
    plt.show()
```

---

```python
plot_evol('Bitcoin',btc,'2019','Close')
```

```python
plot_evol('Bitcoin',btc,'2020','Close')
```

```python
plot_evol('HTG',htg,'2020','Close')
```

```python
plot_evol('ETH',eth,'2020','Close')
```
